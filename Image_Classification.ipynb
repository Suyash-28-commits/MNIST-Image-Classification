```python
from tensorflow.keras.datasets import mnist
import numpy as np

# Load MNIST from TensorFlow (comes in train/test split)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Combine both parts into one full dataset
X_full = np.concatenate((X_train, X_test), axis=0)
y_full = np.concatenate((y_train, y_test), axis=0)

# Flatten images from 28x28 → 784
X_full = X_full.reshape(-1, 28*28).astype("float32")
print("Full dataset shape:", X_full.shape)
print("Full labels shape:", y_full.shape)
```

```python
df = pd.DataFrame(X_full)
df["target"] = y_full
```

```python
#Schochastic Gradient Descent Classifier
#SGDC works on randomness,it shuffles the data
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train,y_train_5)
```

```python
#skewed datasets (i.e., when some classes are much more frequent than others)
#For skewed datasets stratified sampling is required such that proportion of claases in original dataset
#and traing and testing set remains same
```

```python
#Performance measures:
#1.cross_val_score() with StratifiedSampling
from sklearn.model_selection import StratifiedKFold , cross_val_score
skf = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)

cross_val_score(sgd_clf,X,y_5,cv=skf)
```

```python
#cross_val_score() checks how the data works on unseen data
```

```python
#Never-5 classifier
from sklearn.base import BaseEstimator, TransformerMixin
class Never5Classifier(BaseEstimator,TransformerMixin):
  def fit(self,X,y=None):
    return self
  def predict(self,X):
    return np.zeros((len(X),1),dtype=bool)

#this dummy classifier doesn't learn from training data
#It gives only a numpy array of false values
never_5_clf = Never5Classifier()
cross_val_score(never_5_clf,X_train,y_train_5,cv=3,scoring='accuracy')
```

```python
#Accuracy isn't preferred as a valid performance measure here as it is a skewed dataset
```

```python
#Deciding the threshold
y_scores = cross_val_predict(sgd_clf,X_train,y_train_5,cv=3,method='decision_function')
```

```python
y_scores
```

```python
y_scores.shape
```

```python
#Computation of precision and recall of all curves
from sklearn.metrics import precision_recall_curve
precisions , recalls , thresholds = precision_recall_curve(y_train_5,y_scores)
```

```python
#Plotting Precison-recall graph
def plot_precison_vs_recall_graph(precisions , recalls ,thresholds):
  plt.plot(thresholds,precisions[:-1],"b--",label="Precision")
  plt.plot(thresholds,recalls[:-1],"g-",label="Recall")
  plt.xlabel("Threshold")
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
plot_precison_vs_recall_graph(precisions,recalls,thresholds)
```

```python
plt.plot(recalls[:-1],precisions[:-1])
plt.ylabel("Precision")
plt.xlabel("Recall")
plt.grid(True)
plt.show()
```

```python
#Threshold at which Precision is 90%
import numpy as np
threshold_90 = thresholds[np.argmax(precisions>=0.9)]
```

```python
y_pred_90 = (y_scores >= threshold_90)
```

```python
precision_score(y_train_5,y_pred_90)
```

```python
recall_score(y_train_5,y_pred_90)
```

```python
from sklearn.metrics import roc_curve
fpr , tpr , thresholds = roc_curve(y_train_5,y_scores)
```

```python
def plot_roc_curve(fpr,tpr):
  plt.plot(fpr,tpr)
  plt.plot([0,1],[0,1],'k--') #Dashed diagnol line represents a random classifier
  plt.xlabel('False Positive Rate')
  plt.ylabel('Recall')
  plt.title("ROC(Reciever Operating Characteristic) Curve")
  plt.show()
```

```python
plot_roc_curve(fpr,tpr)
```

```python
#Area Under ROC Curve metrics
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5,y_scores)
```

```python
def roc_rf_sgdc(fpr,tpr,fpr_forest,tpr_forest):
  plt.plot(fpr_forest,tpr_forest,"b",label="Random forest")
  plt.plot(fpr,tpr,"r",label="Gradient Descent")
  plt.plot([0,1],[0,1],'k--')
  plt.xlabel("False Positive Rate")
  plt.ylabel("Recall")
  plt.title('Random Forest v/s Gradient Descent')
  plt.legend()
  plt.show()
```
